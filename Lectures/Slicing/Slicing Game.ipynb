```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordsearch import make_grid as mg
```

```python
#%matplotlib inline
%matplotlib notebook
%pylab inline
np.random.seed(5)
```

```python
d = pd.read_excel("./Dictionary.xlsx", sheetname="Sheet2")
```

```python
wcats = d["Category"].unique()
curcat = numpy.random.choice(wcats)
words = d[d["Category"] == curcat]
w = words.head(10)["Word.1"].values
```

```python
#grid = mg(words=["one", "circle", "Hermionie", "Big Brother", "privacy"])
grid = mg(words=w)
puzzle = grid.to_text().upper().split("\n")
puzzle = [list(puzzle[i]) for i in range(len(puzzle))]
p = np.array(puzzle)
```

```python
plt.xkcd()
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)
ax.axis('equal')

ax.set_xlim(-1, 15)
ax.set_ylim(-1, 15)
ax.set_ylim(ax.get_ylim()[::-1])
xticks(np.arange(0, 15, step=1))
yticks(np.arange(0, 15, step=1))
ax.grid(True, axis="both")

plt.title(curcat)
for i in range(15):
    for j in range(15):
        ax.text(i, j, (puzzle[j][i]),size=14)

for i, w in enumerate(w):
    ax.text(20, i, (w),size=14)
```
